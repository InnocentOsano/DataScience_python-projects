### Numpy random distribution assignment

```python
import numpy as np
from numpy import random as rd
```

```python
# 1
coin=np.array([4, 7, 8, 8, 6, 7, 6, 5, 4, 3])
```

```python
# using the normal distribution
#(loc(mean),scale(st.deviation),size)
loc=np.mean(coin)
scale=np.std(coin)
```

```python
loc
```

```python
scale=round(scale,2)
scale
```

```python
nxt100=rd.normal(loc,scale,100)
```

```python
nxt100=np.trunc(nxt100)
```

```python
nxt100
```

```python
# 2
score=rd.normal(70,10,50)
```

```python
score
```

```python
score=np.around(score,2)
score
```

```python
#3
cust=rd.poisson(3,1000)
cust
```

```python
print(np.sum(cust>=7)/len(cust))
```
