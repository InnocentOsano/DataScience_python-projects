 2 write a comment about what you notice when using each of them

```python
cars={"Tesla", "Benz", "Jeep", "Audi", "Volvo"}
print (cars)
```

```python
# using remove ()   # when using remove, if you remove an item not in the list a key  error message pops up.
cars.remove('Tesla')
print(cars)
```

```python
# using discard () # when using discard, if you discard an item not in a set, no key error message pops up
cars.discard('Benz')
print (cars)
```

```python
cars={ "Jeep", "Audi", "Volvo"}
print (cars)
```
