```python
### Simple calculator project 
Project description: you are tasked with creating a simple command-line caluclator using python.
    1 The calculator should be capable of performing the following operations:addition, substraction,multiplication and division.
    2 It should display a menu to the user, prompting them to select the desired operation.
    3 After selecting the operation, the user should input two numbers
    4 The calculator should perfom the chosen operation on the input numbers and display the result to the user.
    5 The user should be able to exit the program gracefully.

```

```python
def calculator ():
    print ('Enter 1 for addition')
    print ('Enter 2 for division')
    print ('Enter 3 for Multiplication')
    print ('Enter 4 for substraction')
    prompt=int(input( 'Enter number: '))
    if prompt==1:
        print ('You selected addition')
        num1=int(input('Enter 1st number'))
        num2=int(input('Enter 2nd number'))
        return num1+num2
    elif prompt==2:
        print( 'You selected division')
        num1=int(input('Enter 1st number'))
        num2=int(input('Enter 2nd number'))
        return num1/num2
    elif prompt==3:
        print('You selected multiplication')
        num1=int(input('Enter 1st number'))
        num2=int(input('Enter 2nd number'))
        return num1*num2
    else:
        print( 'You selected substraction')
        num1=int(input('Enter 1st number'))
        num2=int(input('Enter 2nd number'))
        return num1-num2
```
