```python
### Assignment 2
def highest_even_lowest_odd(numbers):
    evens = [num for num in numbers if num % 2 == 0]
    odds = [num for num in numbers if num % 2 != 0]
    
    if evens and odds:
        highest_even = max(evens)
        lowest_odd = min(odds)
        return highest_even, lowest_odd
    else:
        return None
        
a = [10, 19, 11, 56, 98, 26, 73, 99, 17]
print(highest_even_lowest_odd(a)) 


```
