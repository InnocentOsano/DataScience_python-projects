```python
Programming Task
Create a simple telecom service program using Python classes with the following specifications.
The user should start with a fixed balance of $2000.
The program should display a main menu with the following options
1.Deposit funds
2.Buy airtime
3.Buy data
4.Check balance
5.Exit

If the user selects Buy Airtime or Buy Data, they should be presented with a second-level menu of specific price options 
(e.g., 
1.100, 
2.200, 
3.500, 
4.1000).
After every transaction (deposit, airtime, or data purchase), the userâ€™s balance should be updated and displayed.
Include basic error handling to manage invalid inputs.
Use a class structure with methods such as:
deposit()
by_airtime()
by_data()
show_balance()
menu() (to control the user interface loop)
```

```python
class TelecomService:
    def __init__(self):
        self.balance = 2000  # Starting balance

    def deposit(self):
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount <= 0:
                print(" Deposit amount must be greater than 0.")
            else:
                self.balance += amount
                print(f" Successfully deposited ${amount:.2f}.")
                self.show_balance()
        except ValueError:
            print(" Invalid input! Please enter a number.")

    def buy_airtime(self):
        options = {1: 100, 2: 200, 3: 500, 4: 1000}
        print("\n-- Buy Airtime --")
        for key, value in options.items():
            print(f"{key}. ${value}")

        try:
            choice = int(input("Choose an option: "))
            if choice in options:
                amount = options[choice]
                if self.balance >= amount:
                    self.balance -= amount
                    print(f" Airtime purchase of ${amount} successful.")
                else:
                    print(" Insufficient balance!")
            else:
                print(" Invalid option.")
        except ValueError:
            print(" Please enter a valid number.")
        self.show_balance()

    def buy_data(self):
        options = {1: 100, 2: 200, 3: 500, 4: 1000}
        print("\n-- Buy Data --")
        for key, value in options.items():
            print(f"{key}. ${value}")

        try:
            choice = int(input("Choose an option: "))
            if choice in options:
                amount = options[choice]
                if self.balance >= amount:
                    self.balance -= amount
                    print(f" Data purchase of ${amount} successful.")
                else:
                    print("Insufficient balance!")
            else:
                print(" Invalid option.")
        except ValueError:
            print(" Please enter a valid number.")
        self.show_balance()

    def show_balance(self):
        print(f" Current Balance: ${self.balance:.2f}")

    def menu(self):
        while True:
            print("\n--- Telecom Service Menu ---")
            print("1. Deposit funds")
            print("2. Buy airtime")
            print("3. Buy data")
            print("4. Check balance")
            print("5. Exit")

            try:
                choice = int(input("Enter your choice: "))
                if choice == 1:
                    self.deposit()
                elif choice == 2:
                    self.buy_airtime()
                elif choice == 3:
                    self.buy_data()
                elif choice == 4:
                    self.show_balance()
                elif choice == 5:
                    print(" Exiting. Thank you for using Telecom Service.")
                    break
                else:
                    print(" Invalid choice. Please try again.")
            except ValueError:
                print(" Please enter a valid number.")

       # Run program
if __name__ == "__main__":
    service = TelecomService()
    service.menu()
            
                
                


```
